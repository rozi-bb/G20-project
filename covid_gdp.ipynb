````python
def growth_factor(df, col, new_col):
    growth = [0]
    for x in range(1, (len(df))):
        if df[col][x] == 0 and df[col][x-1] == 0:
          grow = 1
          growth.append(grow)
        elif df[col][x] > 0 and df[col][x-1] == 0:
          grow = 0.99
          growth.append(grow)
        elif df[col][x] == 0 and df[col][x-1] > 0:
          grow = 0
          growth.append(grow)
        else:
          grow = df[col][x] / df[col][x-1]
          growth.append(grow)
    df[new_col] = growth
````

````python
growth_factor(df, 'New Recovered', 'Growth Factor of New Recovered')
growth_factor(df, 'New Active Cases', 'Growth Factor of New Active Cases')
````

## Hypothesis Testing

```
Interpretasi :
H0: Tidak ada hubungan antara kasus Covid19 dengan pertumbuhan ekonomi di Indonesia.
H1: Terdapat hubungan antara kasus Covid19 dengan pertumbuhan ekonomi di Indonesia.
```

````python
corr_1 = df['Total Cases'].corr(df['gdp_growth'], method ='pearson')
round(corr_1, 3)
````

````python
corr_2 = df['Total Cases'].corr(df['gdp'], method ='pearson')
round(corr_2, 3)
````

````python
corr_3 = df['Case Fatality Rate'].corr(df['gdp_growth'], method ='pearson')
round(corr_3, 3)
````

````python
from scipy import stats
res = stats.pearsonr([1, 2, 3, 4, 5], [10, 9, 8, 7, 10])
res
````

````python
#the Pearson's Correlation test
from scipy.stats import pearsonr
stat, p_value = pearsonr(df['Total Cases'], df['gdp_growth'])
print('stat=%.3f, pvalue=%.3f' % (stat, p_value))
if p_value > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')
````

**Since our p-value is less than `0.05`, so we reject the null hypothesis** and we can conclude that, there is correlation between hypothesis.
